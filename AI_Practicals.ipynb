# **Practical 2: Variables, Data Types, Operators, Loop & Iterations in Python**

```python
print("=== Data Type ===")
a = 10
b = 20.0
c = "abc"
d = True
print("Type of a : ", type(a))
print("Type of b : ",type(b))
print("Type of c : ",type(c))
print("Type of d : ",type(d))


print("=== Arithnatic operator ===")
a = int(input("Enter first number : "))
b = int(input("Enter second number : "))
print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)


print("=== Comparision operator ===")
a = int(input("Enter first number : "))
b = int(input("Enter second number : "))
print("a < b : ", a < b)
print("a > b : ", a > b)
print("a != b : ", a != b)



print("=== Logical operator ===")
A = True
B = False
print("A = True, B = False")
print("A and B : ", A and B)
print("A or B : ", A or B)
print("not A : ", not A)


print("=== Assignment operator ===")
a = int(input("Enter first number : "))
b = int(input("Enter second number : "))
a += 2
print("a += 2 : ", a)
b -= 2
print("b -= 2 : ",b)
a *= b
print("a *= b : ",a)
a /= b
print("a /= b : ",a)


print("=== print N number using for loop ===")
n = int(input("Enter a number : "))

for i in range(1, n+1):
  print(i)



print("=== Factorail of N number using while loop ==")
a = int(input("Enter a number : "))
fact = 1
i = 1
while i <= a:
  fact *= i
  i += 1
print("Factorial : ", fact)
```
